```python
from scipy import spatial
from sent2vec.vectorizer import Vectorizer

sentences = [
    " I love to eat ice creams its just so good.",
    " I don't have a girlfiend and I really need one.",
    " I love chocoloate flavour in ice creams."
]
```

Using the pre trained model "distilbert-base-multilingual-cased" for training the model with the weights that model has.

```python
vectorizer = Vectorizer(pretrained_weights= 'distilbert-base-multilingual-cased')
```

Vectorizing the sentences and removing the stopwords 

```python
vectorizer.run(sentences,remove_stop_words=[['the'],['so'],['and'],['i']] , add_stop_words=[])
vectors_w2v = vectorizer.vectors
```

```python
vectors_w2v
```

Distance between the sentence usig cosine similarity

```python


dist_1 = spatial.distance.cosine(vectors_w2v[0], vectors_w2v[1])
dist_2 = spatial.distance.cosine(vectors_w2v[0], vectors_w2v[2])

print('dist_1: {}'.format(dist_1))
print('dist_2: {}'.format(dist_2))
```
